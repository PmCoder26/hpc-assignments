```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("result.txt")
```

```python
plt.figure(figsize=(8,5))
plt.plot(df.N, df.SEQ, "b*-", label="Sequential")
plt.plot(df.N, df.PAR, "g*-", label="Parallel")
plt.xlabel("Input Size N")
plt.ylabel("Time (seconds)")
plt.title("Sequential vs Parallel")
plt.legend()
plt.grid()
plt.show()
```

```python
# Graph 2: Speedup

plt.figure(figsize=(8,5))
plt.plot(df.N, df.SPEEDUP,"r*-", label="Speedup")
plt.xlabel("Input Size N")
plt.ylabel("Speedup (SEQ / PAR)")
plt.title("Speedup Analysis")
```
